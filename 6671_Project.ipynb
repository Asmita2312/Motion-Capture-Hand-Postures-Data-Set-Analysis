```
import pandas as pd

filename = 'Postures.csv'
data = pd.read_csv(filename)

print(data.shape)
print(data.columns)
data.head(10)
```

```
import numpy as np
data = data.drop(data.index[0])

#Since missing values are represented by '?', in order to do pre-processing, 
#first I am converting all '?' to Nan that is numeric type

data['X0'] = pd.to_numeric(data['X0'], errors='coerce')
data['Y0'] = pd.to_numeric(data['Y0'], errors='coerce')
data['Z0'] = pd.to_numeric(data['Z0'], errors='coerce')
data['X1'] = pd.to_numeric(data['X1'], errors='coerce')
data['Y1'] = pd.to_numeric(data['Y1'], errors='coerce')
data['Z1'] = pd.to_numeric(data['Z1'], errors='coerce')
data['X2'] = pd.to_numeric(data['X2'], errors='coerce')
data['Y2'] = pd.to_numeric(data['Y2'], errors='coerce')
data['Z2'] = pd.to_numeric(data['Z2'], errors='coerce')
data['X3'] = pd.to_numeric(data['X3'], errors='coerce')
data['Y3'] = pd.to_numeric(data['Y3'], errors='coerce')
data['Z3'] = pd.to_numeric(data['Z3'], errors='coerce')
data['X4'] = pd.to_numeric(data['X4'], errors='coerce')
data['Y4'] = pd.to_numeric(data['Y4'], errors='coerce')
data['Z4'] = pd.to_numeric(data['Z4'], errors='coerce')
data['X5'] = pd.to_numeric(data['X5'], errors='coerce')
data['Y5'] = pd.to_numeric(data['Y5'], errors='coerce')
data['Z5'] = pd.to_numeric(data['Z5'], errors='coerce')
data['X6'] = pd.to_numeric(data['X6'], errors='coerce')
data['Y6'] = pd.to_numeric(data['Y6'], errors='coerce')
data['Z6'] = pd.to_numeric(data['Z6'], errors='coerce')
data['X7'] = pd.to_numeric(data['X7'], errors='coerce')
data['Y7'] = pd.to_numeric(data['Y7'], errors='coerce')
data['Z7'] = pd.to_numeric(data['Z7'], errors='coerce')
data['X8'] = pd.to_numeric(data['X8'], errors='coerce')
data['Y8'] = pd.to_numeric(data['Y8'], errors='coerce')
data['Z8'] = pd.to_numeric(data['Z8'], errors='coerce')
data['X9'] = pd.to_numeric(data['X9'], errors='coerce')
data['Y9'] = pd.to_numeric(data['Y9'], errors='coerce')
data['Z9'] = pd.to_numeric(data['Z9'], errors='coerce')
data['X10'] = pd.to_numeric(data['X10'], errors='coerce')
data['Y10'] = pd.to_numeric(data['Y10'], errors='coerce')
data['Z10'] = pd.to_numeric(data['Z10'], errors='coerce')
data['X11'] = pd.to_numeric(data['X11'], errors='coerce')
data['Y11'] = pd.to_numeric(data['Y11'], errors='coerce')
data['Z11'] = pd.to_numeric(data['Z11'], errors='coerce')
#data.head(10)
```

```
data.head(5)
```

```
#Now I am replacing all Nan values with mean of corresponding columns
new_data = data.apply(lambda x: x.fillna(x.mean()),axis=0)
new_data
```

```
#Since the data set was sorted in ascending order, I have used sample function to shuffle rows in dataset,
# So that while splitting dataset into train and test data, it will have values for each posture class
new_data = new_data.sample(frac=1)
new_data.head(5)
```

```
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score

#Here I have separated feature column and label column. 
#Since 'User' column was not giving any information, I have dropped that column 

X = new_data.drop(columns = ['Class','User'])
Y = new_data[['Class']]
print(X)
```

```
print(Y)
```

```
train_x, test_x, train_y, test_y = train_test_split(X, Y, train_size=0.7, random_state=25)
```

```
print(train_x.shape)
print(train_y.shape)
```

```
# For trying different variants in network :
#I have used different shape for each layer, also I tried various activation functions and epoch values

import tensorflow as tf
import keras
from keras import optimizers
from keras.models import Sequential
from keras import layers

model = Sequential()
model.add(layers.Dense(45, activation=tf.nn.sigmoid, input_shape=(36, )))
model.add(layers.Dense(36, activation=tf.nn.sigmoid))
model.add(layers.Dense(10, activation=tf.nn.sigmoid))
model.add(layers.Dense(25, activation=tf.nn.sigmoid))
model.add(layers.Dense(10, activation=tf.nn.softmax))

model.summary()
```

```
model.compile(optimizer=tf.train.AdamOptimizer(), 
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.fit(train_x, train_y, epochs=5)
```

```
#Here I have changed the layer shape and activation functions

model1 = Sequential()
model1.add(layers.Dense(120, activation=tf.nn.relu, input_shape=(36, )))
model1.add(layers.Dense(45, activation=tf.nn.relu))
model1.add(layers.Dense(32, activation=tf.nn.relu))
model1.add(layers.Dense(60, activation=tf.nn.softmax))

model1.summary()
```

```
model1.compile(optimizer=tf.train.AdamOptimizer(), 
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```
model1.fit(train_x, train_y, epochs=15)
```

```
#Since with above actiavtion function and layers, I got expected accuracy, 
#here I have changed only epoch, keeping rest everying same

model1.fit(train_x, train_y, epochs=25)
```

```
# Here I've created a loop that breaks a dataset and put yeild inside loop to make it as a generator
def chunks_X(X):
  for g, df in X.groupby(np.arange(len(X)) // 350):
    yield df
chunk_X = chunks_X(X)

# Printing all chunks with next() function
for j in chunk_X:
    batch_x = next(chunk_X)
    print(batch_x)
```

```
# Repeating same steps of generator for labels

def chunks_Y(Y):
  for g, df in Y.groupby(np.arange(len(Y)) // 350):
    yield df
chunk_Y = chunks_Y(Y)

for j in chunk_Y:
    batch_y = next(chunk_Y)
    print(batch_y)    
```

```
print(type(batch_x))
print(type(batch_y))
```

```
# Here I am randomly updating 10 values in each batch with 0 as an augmentation technique
#Since last batch has shape (45,36), I have used if-else to take care of batch shape
#In the end I am concatenating all batches to create new data frame which has 0 at random indices
df_n = pd.DataFrame(columns=['X0','Y0','Z0','X1','Y1','Z1','X2','Y2','Z2','X3','Y3','Z3','X4','Y4','Z4','X5','Y5','Z5',
                              'X6','Y6','Z6','X7','Y7','Z7','X8','Y8','Z8','X9','Y9','Z9','X10','Y10','Z10','X11','Y11','Z11'])
df_y = pd.DataFrame(columns=['Class'])

chunk_X = chunks_X(X)
for j in chunk_X:
  batch_x = next(chunk_X)
  for i in range(0,10):
    if(batch_x.shape == (45,36)):
      n = np.random.randint(1,45)
      m = np.random.randint(1,36)
    else:
      n = np.random.randint(1,350)
      m = np.random.randint(1,36)
  batch_x.iat[n,m] = 0
  df_interpolated = batch_x
  df_n = pd.concat([df_n,df_interpolated])


chunk_Y = chunks_Y(Y)
for i in chunk_Y:
  batch_y = next(chunk_Y)
  df_y = pd.concat([df_y,batch_y])
```

```
model2 = Sequential()
model2.add(layers.Dense(120, activation=tf.nn.relu, input_shape=(36, )))
model2.add(layers.Dense(45, activation=tf.nn.relu))
model2.add(layers.Dense(32, activation=tf.nn.relu))
model2.add(layers.Dense(60, activation=tf.nn.softmax))

model2.summary()
```

```
model2.compile(optimizer=tf.train.AdamOptimizer(), 
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```
history = model2.fit_generator((df_n,df_y),
                    steps_per_epoch = 45, 
                    epochs=15)
```
